```python
import pandas as pd
```

```python
df1 = pd.read_csv('20220628scopus.csv').drop(['Link','Authors with affiliations'],axis=1)
df1.head()
```

```python
df1.columns = ['authors', 'title', 'year', 'journal', 'affiliations', 'abstract']
df1.head()
```

```python
df2 = pd.read_csv('20220628DOAJ.csv').drop('Unnamed: 0',axis=1)
df2.head()
```

```python
intersection = pd.merge(df1['title'].str.lower() , df2['title'].str.lower())
intersection
```

```python
overlap_index = intersection.shape[0] / min(df1.shape[0],df2.shape[0])
overlap_index
```

```python
df = pd.merge(df1,df2,how='outer')
df.shape
```

```python
df['title'] = df['title'].str.lower()
```

```python
df = df.drop_duplicates(subset=['title']).reset_index(drop=True)
df.shape
```

```python
df['title'] = df['title'].str.title()
```

```python
df.to_csv('merged_dataset.csv')
```
